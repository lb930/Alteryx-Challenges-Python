In this week's challenge, we have the results from the New York City Public School standardized math exams from 2013 to 2019.  Examining only 8th grade students, find the percentage of female students that achieved a Level 4 score for each district in the years 2013 and 2019.  Then rank the districts according to the biggest improvement in this percentage.

Data Source: https://infohub.nyced.org/reports/academics/test-results

Original challenge: https://community.alteryx.com/t5/Weekly-Challenge/Challenge-219-Who-Runs-The-World-Girls/td-p/584238

```python
import pandas as pd
```

```python
df = pd.read_excel(r'.\input_219.xlsx')
df
```

```python
df = df[(df['Grade'] == '8') & (df['Category'] == 'Female') & ((df['Year'] == 2013) | (df['Year'] == 2019))]
df
```

```python
df = df.pivot(index = 'District', columns = 'Year', values = '% Level 4').reset_index()

# Renames index column from Year to None
df.rename_axis([None], axis = 1, inplace = True)
df.head()
```

```python
df['Difference'] = df[2019].to_numpy() - df[2013].to_numpy()
df.head()
```

```python
df.sort_values(by = 'Difference', inplace = True, ascending = False)
df.head()
```
