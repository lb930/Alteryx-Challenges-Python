## Challenge #9: Analytics Ranking

Original challenge: https://community.alteryx.com/t5/Weekly-Challenge/Challenge-9-Analytics-Ranking/td-p/36736

The objective is to determine the top 5 ranking based on the count, however since there are multiple rows with same count (similar to a round of golf) multiple people can be in the same place (Rank) if they have the same score.

```python
import pandas as pd
```

```python
#read file and explore data

df = pd.read_csv(".Challenge 009/challenge_9_input.csv")
df
```

```python
#sort DataFrame descending by count column to find the person with the highest count

df.sort_values(by = "count", ascending = False, inplace = True)
```

```python
#create a dense rank for each person

df["rank"] = df["count"].rank(method = "dense", ascending = False)
df
```

```python
#exclude ranks which are greater than 5

df[df["rank"]<=5.0]
```
