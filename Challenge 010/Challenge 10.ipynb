```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Luisa Bez\Documents\Python Scripts\Alteryx challenges\challenge_10_input.csv", parse_dates=["TIMESTAMP", "Time_Now"])
df.head()
```

```python
df["Datediff"] = (df["Time_Now"] - df["TIMESTAMP"])
df.head()
```

```python
def parse_date_strings(dataframe):
    dataframe = dataframe["Datediff"]
    return [dataframe.components.days, dataframe.components.hours, dataframe.components.minutes, dataframe.components.seconds]
```

```python
df[["Days", "Hours", "Minutes","Seconds"]] = df.apply(parse_date_strings, axis = 1, result_type = "expand")
df.drop(["Datediff"], axis = 1, inplace = True)
df
```
