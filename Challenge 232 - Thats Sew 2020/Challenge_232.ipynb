## Challenge 232 - That's Sew 2020

In the attached datasets, we have the number of masks created by a local community known as Bastille St. Below that, we also have the number of masks created by local companies and independent contractors for the same community. Use this data to find the total number of masks produced by each group for each mask size and the total number of masks produced by each group.

Original challenge: https://community.alteryx.com/t5/Weekly-Challenge/Challenge-232-That-s-Sew-2020/td-p/630342

```python
import pandas as pd
```

```python
bastille = pd.read_csv('.\input1.csv', encoding = 'ISO-8859-1')
bastille
```

```python
butterfly = pd.read_csv('.\input2.csv')
butterfly
```

```python
bastille['Source'] = 'Bastille'
bastille = bastille.groupby(by = 'Source').sum(['Large', 'Medium', 'Small']).reset_index()
bastille
```

```python
bastille['Total'] = bastille.sum(axis = 1)
bastille
```

```python
butterfly['Total'] = butterfly.sum(axis = 1)
butterfly
```

```python
df = pd.concat([bastille, butterfly.rename(columns = {'Name': 'Source'})])
df
```

```python
df.append(df.sum(numeric_only=True), ignore_index=True)
```
