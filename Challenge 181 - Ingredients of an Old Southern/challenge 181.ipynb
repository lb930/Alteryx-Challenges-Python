## Challenge #181: Ingredients of an Old Southern

Original challenge: https://community.alteryx.com/t5/Weekly-Challenge/Challenge-181-Ingredients-of-an-Old-Southern/td-p/464549

Using the provided URL, download and parse the data in the recipe for an Old Southern. Create a list of the cocktail ingredients and measurements.

```python
import pandas as pd
import requests, io
import json
```

```python
df = pd.read_csv(".Challenge 181/challenge_181_input.csv")
df
```

```python
#Get data in JSON format from URL
urlData = requests.get(df.loc[0]["Field1"]).content
urlData
```

```python
#Convert bytes into dictionary
jsonResponse = json.loads(urlData.decode('utf-8'))
jsonResponse
```

```python
#Acces values in "drinks" key and add them to a DataFrame
df2 = pd.DataFrame(jsonResponse["drinks"][0], index = [0])
df2
```

```python
#Pivot DataFrame to enable easier filtering
df2 = pd.melt(df2)
df2
```

```python
#Exclude null values
df2 = df2[df2["value"].notnull()]
df2.head(20)
```

```python
#Keep rows which contain ingredients and measures
ingredient = df2[df2["variable"].str.contains("Ingredient")]
measure = df2[df2["variable"].str.contains("Measure")]
```

```python
#Extract number from variable to use as key for join
ingredient["id"] = ingredient["variable"].str.extract("(\d)")
measure["id"] = measure["variable"].str.extract("(\d)")
```

```python
ingredient
```

```python
measure
```

```python
#Join ingredient and measure DataFrame to obtain final DataFrame
result = pd.merge(ingredient, measure, how = "inner", on = "id")
result
```

```python
#Drop and rename columns according to output csv file
result.drop(["variable_x", "variable_y", "id"], axis = 1, inplace = True)
result.columns = ["Ingredients", "Measure"]
result
```
