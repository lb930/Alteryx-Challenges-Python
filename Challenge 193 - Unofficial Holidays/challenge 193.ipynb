## Challenge #193: Unofficial Holidays
Original challenge: https://community.alteryx.com/t5/Weekly-Challenge/Challenge-193-Unofficial-Holidays/td-p/500029

List the unofficial holidays provided in the input in chronological order. 

```python
import pandas as pd
```

```python
#read files and exlore data
df = pd.read_csv(".Challenge 193/challenge_193_input.csv")
df
```

```python
#remove leading and trailing whitespace
df = df.apply(lambda x: x.str.strip())
df
```

```python
#change date format from string to datetime so it can be sorted
df["date"] = pd.to_datetime(df["day"], format = "%B %d")
```

```python
df.sort_values(by = "date", inplace = True)
df
```

```python
#delete date column
del df["date"]
```
